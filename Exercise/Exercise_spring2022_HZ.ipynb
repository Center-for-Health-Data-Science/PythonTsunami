### Solution

```python
#1:
stroke['stroke'].value_counts()
```

```python
#3:
cts = pd.DataFrame(stroke[['stroke','work_type']].value_counts())
```

Congrats, this is quite close to what we need. Now we need to turn these counts into fractions by dividing them by the total per outcome. But first, let's make this into a proper dataframe. You can probably see that some columns stretch over several rows (at least in our example). We'll correct that by resetting the index and assign some column names.

```python
cts = cts.reset_index()
cts.columns = ['stroke', 'work_type', 'count']
```

Onwards to the fractions calculation. The fraction of people who had a stroke and work in the private sector is the count of them divided by the sum of all people who had a stroke. Use groupby and calculate the totals.

### Solution

```python
cts.groupby('stroke').sum()
```

Great! Though we need to divide every row by the correct sum, i.e. in rows that refer to stroke == 1 the counts need to be divided by the sum of all stroke == 1, not stroke == 0. We will therefore use a neat little trick called transform to produce an output (series or dataframe) that has the same length as the original input. We'll also select the counts column (you can run without that and see what it does to the work_type column).

```python
cts.groupby('stroke')['count'].transform('sum')
```

And here we have our desired column:

```python
cts['perc'] = cts['count'] / cts.groupby('stroke')['count'].transform('sum')
```

Inspect the resulting dataframe. This is the format we needed for the histplot code to work. Now what's left is to adapt the code we found to our data. 

### Solution

```python
#change the variables to our data:
ax = sns.histplot(data=cts, x='stroke', hue='work_type', weights = 'perc', multiple='stack', shrink=0.8)
ax.set_ylabel('percentage')
# Fix the legend so it's not on top of the bars.
legend = ax.get_legend()
legend.set_bbox_to_anchor((1, 1))
```

### Part 4: Imputation

__Proceed with extreme caution!__

Imputation is the replacement of missing values with likely values. Depending on your dataset this can be a good or a very bad idea. For some data types, the average of all observations is a good guess for a missing observation. However, especially when it comes to categorical data, imputation can produce results that are not meaningful. 

For example, in the cirrhosis data we have a column called `Drug`. Have a look at the data type of this column. how many different values are there? What do they mean?



```python
cirrhosis.info()
cirrhosis['Drug'].value_counts()
```

This column is about whether or not the person was treated with the drug or the placebo. Unlike measurements like i.e. the amount of cholesterol we have no good way to guess which treatment the person received. 
Replacing missing values in this column with the most common value (which is that they did get the drug) will produce extremely __wrong data__ and lead you to __wrong conclusions__.
