## Exercise 3

Make a new dataframe with the date, number of cases and number deaths for Denmark.

Which country has most daily deaths per 100.00?

Make the country code column lower-case

## Exercise 6

What was the median number of daily cases per country?

```python
df.groupby(by='countriesAndTerritories')['cases'].median()
```

On average, how many cases are there for each month in each country?

```python
df.groupby(['countriesAndTerritories', 'year', 'month'])['cases'].mean()
```

How many days where there without deaths in each country?

### Select days with no deaths

```python
no_death = df.loc[df['deaths']==0]
no_death.head()
```

### Use `.count()` to count rows

```python
no_death.groupby('countriesAndTerritories').count()
```

### Reduce to only one column

```python
no_death.groupby('countriesAndTerritories').count()['dateRep']
```

## Exercise 7

Calculate how many were infected and how many died daily, on average, in each country in Europe?

```python
df.loc[df['continentExp']=='Europe'].groupby('countriesAndTerritories').agg({'cases':'mean', 'deaths':'mean'})
```

What was the highest and the lowest number of infected per month, per country?

```python
df.groupby(['year', 'month', 'countriesAndTerritories']).agg({'cases':[max, min]})  # find anything curious in the results?
```
