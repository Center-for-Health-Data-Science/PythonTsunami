# Matplotlib

Python's plotting library.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Prepare the data

```python
x = np.arange(1,11)
y1 = np.array([4,5,6,7,20,33,44,55,66,77])
y2 = np.random.random(10) * 100
y3 = np.random.random(10) * 100
y4 = np.random.random(10) * 100
```

Do simple plot

```python
plt.plot(x,y1) # note that plt.plot(x=x, y=y1) doesn't work
```

Change plot size

```python
fig, ax = plt.subplots(figsize=(10,7))  # size in inches
ax.plot(x,y1)
```

Plot multiple subplots in the same figure

```python
fig, ax_list = plt.subplots(nrows=2, ncols=2, figsize=(10,7))  # size in inches
ax_list[0,0].plot(x,y1)
ax_list[0,1].plot(x,y2)
ax_list[1,0].plot(x,y3)
ax_list[1,1].plot(x,y4)
```

```python
from IPython.display import Image
Image("images/matplotlib_fig_anatomy.png")
```

 - The Figure is the final image that may contain 1 or more Axes.

 - The Axes represent an individual plot (don't confuse this with the word "axis", which refers to the x/y axis of a plot).

Plot different data in the sample plot, just play around :)

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,7))

ax.plot(x,y1, color='#a6cee3', label='y1')
ax.plot(x,y2, color='blue', label='y2', linewidth=5)
ax.plot(x,y3, color='green', label=r'$y_3$', linestyle='--')
ax.plot(x,y4, color='black', label='y4', marker='o', markersize=10)

ax.legend(loc='upper left', fontsize='12')  # define the legend position with loc, and the font size


ax.set_xlabel('x-axis', fontsize=12)  # define the label for the x axis and the font size
ax.set_ylabel('y-axis', fontsize=12)  # define the label for the y axis and the font size

ax.set_title('My first plot', fontsize=16)  # define the plot title and the font size

ax.set_xlim(0,5)  # define the range for the x axis
ax.set_ylim(0,100)  # define the range for the y axis


ax.grid(axis='y')

ax.spines['top'].set_visible(False)  # make the top spines invisible
ax.spines['right'].set_visible(False)  # make the right spines invisible

ax.axhline(50, color='grey', linestyle=':', linewidth=5)   # draw a grey, dashed (linestyle) horizontal line
ax.axvline(2.5, color='orange', linestyle=':', linewidth=5)  # draw a orange, dashed (linestyle) vertical line
```

### Scatter plots

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,7))
ax.scatter(x,y1)
```

### Histogram

```python
# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,7))
ax.hist(x, bins=50)
```

### Boxplots

```python
all_data = np.array([np.random.normal(1, 2, size=100),
                     np.random.normal(-1, 2, size=100),
                     np.random.normal(3, 2, size=100)]).transpose()

labels = ['x1', 'x2', 'x3']

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))

# rectangular box plot
bplot1 = ax.boxplot(all_data,
                   vert=True,  # vertical box alignment
                   patch_artist=True,  # fill with color
                   labels=labels)  # will be used to label x-ticks
```
