```python
# select backend explicitly
import pandas as pd
# pd.options.plotting.backend = "plotly" # set plotly as plotting backend
pd.options.plotting.backend = "matplotlib" # revert back
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=list("ABCD"))
df = df.cumsum()
```

```python
df.head()
```

```python
df.plot()
```

```python
df.plot.scatter(x='A', y='B')
```

```python
df.iloc[5].plot.bar()
```

```python
df.iloc[5].plot.barh()
```

```python
df.plot.hist()
```

```python
df.plot.hist(bins=20)
```

```python
df.plot.box()
```

```python
df.iloc[800:].plot.area(x='A', y='B')
```
